```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('Auto.csv')
dataset.head()
```

```python
plt.figure(figsize=(9,6))
```

```python
plt.plot(dataset.index, dataset['mpg'], label='MPG', color='blue')
plt.xlabel('Index')
plt.ylabel('Miles Per Gallon (MPG)')
plt.title('MPG Trend Across Cars')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.title("Distribution of MPG")
plt.hist(dataset.mpg,bins=20, color='skyblue', edgecolor='black')
plt.xlabel('MPG')
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(x='cylinders',y='mpg', data = dataset)
plt.title('MPG by cylinder count')
plt.show()
```

```python
sns.pairplot(dataset[['mpg', 'horsepower', 'weight', 'acceleration']])
plt.suptitle('Pair Plot of Auto Features', y=1.02)
plt.show()
```

```python
sns.heatmap(dataset.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python
cylinder_counts = dataset['cylinders'].value_counts()
labels = cylinder_counts.index
sizes = cylinder_counts.values
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Cars by Cylinder Count')
plt.axis('equal') 
plt.figure(figsize=(11,8)) 
plt.show()
```
